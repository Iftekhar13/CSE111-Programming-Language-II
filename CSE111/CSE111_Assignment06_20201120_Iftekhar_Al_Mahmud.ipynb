```python
#Assignment06
#Task1
class Student():
  id = 0
  def __init__(self, n, dept, a, cg):
    self.name = n
    self.department = dept
    self.age = a
    self.cgpa = cg
    Student.id = Student.id + 1
    self.id = Student.id
  def get_details(self):
    print('ID:',self.id)
    print('Name:',self.name)
    print('Department:',self.department)
    print('Age:',self.age)
    print('CGPA:',self.cgpa)
  @classmethod
  def from_String(cls, details):
    n, dept, a, cg = details.split('-')
    ob = cls(n, dept, a, cg)
    return ob

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
#5
print('5. class variables are the one that has values assigned in the class declaration and is not linked with any particular object. Instance variables are variables that are assigned with values within class methods and is linked with the objects of the class.')
#6
print('6. Instance methods require an instance of the class and needs to be accessed via self. Class methods do not need any intance of the class and needs to accessed using cls.')
```

```python
#Task2
class Assassin():
  num = 0
  def __init__(self, n, r):
    self.name = n
    self.rate = r
    Assassin.num = Assassin.num + 1
  @classmethod
  def failureRate(cls, n, r):
    ob = cls(n, 100-r)
    return ob
  @classmethod
  def failurePercentage(cls, n, r):
    ob = cls(n, 100-r)
    return ob
  def printDetails(self):
    print('Name:',self.name)
    print('Success rate:',self.rate)
    print('Total number of Assassin:', Assassin.num) 



john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#Task5
class Employee():
  def __init__(self, n, wp):
    self.name = n
    self.workingPeriod = wp
  @classmethod
  def employeeByJoiningYear(cls, name, year):
    cls.name = name
    cls.year = 2021 - year
    ob = cls(name, cls.year)
    return ob
  @staticmethod
  def experienceCheck(ex, g):
    if ex < 3 and g == 'female':
      return 'She is not experienced'
    elif ex < 3 and g == 'male':
      return 'He is not experienced'
    elif ex >= 3 and g == 'female':
      return 'She is experienced'
    elif ex >= 3 and g == 'male':
      return 'He is experienced'


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#Task7
class Cat():
  Number_of_cats = 0
  def __init__(self, c, b):
    self.color = c
    self.behavior = b
    Cat.Number_of_cats = Cat.Number_of_cats + 1
  @classmethod
  def no_parameter(cls, c = 'White', b = 'sitting'):
    ob = cls(c, b)
    return ob
  @classmethod
  def first_parameter(cls, c, b = 'sitting'):
    ob = cls(c, b)
    return ob
  @classmethod
  def second_parameter(cls, b, c = 'Grey'):
    ob = cls(c, b)
    return ob
  def changeColor(self, c):
    self.color = c
  def printCat(self): 
    print(self.color,'cat is',self.behavior)


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task8
class Cylinder():
  radius = 5
  height = 18
  def __init__(self, n1, n2):
    self.num1 = n1
    self.num2 = n2
    print('Default radius=',Cylinder.radius,'and height=',Cylinder.height)
    Cylinder.radius = self.num1
    Cylinder.height = self.num2
    print('Updated radius=',self.num1,'and height',self.num2)
  @staticmethod
  def volume(n1, n2):
    import math
    print('Volume:',math.pi * float(Cylinder.radius**2) * float(Cylinder.height))
  @staticmethod
  def area(n1, n2):
    import math
    print('Area:',2 * math.pi * float(Cylinder.radius) * float(Cylinder.height) + 2 * math.pi * float(Cylinder.radius**2))
  @classmethod
  def swap(cls, n1, n2):
    ob = cls(n2, n1)
    return ob
  @classmethod
  def changeFormat(cls, info):
    radius,height = info.split('-')
    ob = cls(float(radius), float(height))
    return ob

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task10
class SultansDine():
  total_num = 0
  total_sell = 0
  lst = []
  def __init__(self, branch):
    self.branch = branch
    SultansDine.total_num = SultansDine.total_num + 1
    SultansDine.lst.append(branch) 
  def sellQuantity(self, quantity):
    if quantity < 10:
      self.quantity = quantity * 300
    elif quantity < 20:
      self.quantity = quantity * 350
    else:
      self.quantity = quantity * 400
    self.Branch_sell = self.quantity
    SultansDine.lst.append(self.Branch_sell)
    SultansDine.total_sell = SultansDine.total_sell + self.Branch_sell
    self.sell_percentage = (self.Branch_sell / SultansDine.total_sell) * 100
  def branchInformation(self):
    print('Branch Name:', self.branch)
    print('Branch Sell:', self.Branch_sell,'Taka')
  @classmethod
  def details(cls):
    print('Total Number of branch(s):',SultansDine.total_num) 
    print('Total Sell:',SultansDine.total_sell)

    for x in range(0, len(SultansDine.lst), 2):
      sell_percentage = (SultansDine.lst[x+1] / SultansDine.total_sell) * 100
      new_sell_percentage = str(round(sell_percentage, 2))
      print('Branch Name:',SultansDine.lst[x])
      print('Branch Sell:',SultansDine.lst[x+1],'Taka')
      print("Branch consists of total sell's:",new_sell_percentage,'%')

SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
